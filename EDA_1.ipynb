### Tampilkan Data Nilai Lokasi tapi tidak Location Other

```python
ax =df.query('Location != "Other"') \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean', 'count']) \
    .query('count >= 10') \
    .sort_values('mean')['mean'] \
    .plot(kind='barh', figsize=(12, 5), title="Avarega Coast Speed by Location Coaster")

ax.set_xlabel("Average Coaster Speed")
plt.show()
```
